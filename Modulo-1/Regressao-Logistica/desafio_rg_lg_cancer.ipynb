```
from sklearn.datasets import load_breast_cancer
import pandas as pd
pd.set_option('display.max_columns', 30)
dados = load_breast_cancer()
```

```
x = pd.DataFrame(dados.data, columns=[dados.feature_names])
y = pd.Series(dados.target)
```

```
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.model_selection import GridSearchCV

#Definindo valores para serem testado em LogisticRegression
valores_C = np.array([0.1, 0.5, 1.0, 2.0, 3.0, 100.0])
regularizacao = ['l1', 'l2']
valores_grid = {'C': valores_C, 'penalty' : regularizacao}

#Criacao do Modelo
modelo = LogisticRegression(solver='liblinear', max_iter=1000)

#Criandos os grids
grid_regressao_logistica = GridSearchCV(estimator = modelo, param_grid= valores_grid, cv=5)
grid_regressao_logistica.fit(x,y)
print('Melhor Acuraria', grid_regressao_logistica.best_score_)
print('Melhor Parametro C', grid_regressao_logistica.best_estimator_.C)
print('Melhor Regularizacao', grid_regressao_logistica.best_estimator_.penalty)
```

```
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.model_selection import GridSearchCV

#Definindo valores para serem testado em LogisticRegression
valores_C = np.array([70, 80, 85, 88, 90, 92, 94, 95, 96, 97, 98, 105, 108, 110, 115, 120])
valores_grid = {'C': valores_C}

#Criacao do Modelo
modelo = LogisticRegression(solver='liblinear', max_iter=1000, penalty='l1')

#Criandos os grids
grid_regressao_logistica = GridSearchCV(estimator = modelo, param_grid= valores_grid, cv=5)
grid_regressao_logistica.fit(x,y)
print('Melhor Acuraria', grid_regressao_logistica.best_score_)
print('Melhor Parametro C', grid_regressao_logistica.best_estimator_.C)
print('Melhor Regularizacao', grid_regressao_logistica.best_estimator_.penalty)
```
