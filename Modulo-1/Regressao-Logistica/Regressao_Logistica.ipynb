```
import pandas as pd
pd.set_option('display.max_columns', 64)
pd.set_option('display.max_rows', 64)
dados = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Modulo-1/Regressao-Logistica/datasets_27954_35620_Data_train_reduced.csv')
dados.head()
```

```
#Contabilizando dados faltantes
faltantes = dados.isnull().sum()
faltantes_porcentual = dados.isnull().sum() / len(dados['Product.ID']) * 100
print(faltantes_porcentual)
```

```
#retirando dados missing > 20%
dados.drop('q8.20', axis=1, inplace=True)
dados.drop('q8.18', axis=1, inplace=True)
dados.drop('q8.17', axis=1, inplace=True)
dados.drop('q8.8', axis=1, inplace=True)
dados.drop('q8.9', axis=1, inplace=True)
dados.drop('q8.10', axis=1, inplace=True)
dados.drop('q8.2', axis=1, inplace=True)
dados.drop('Respondent.ID', axis=1, inplace=True)
dados.drop('Product', axis=1, inplace=True)
dados.drop('q1_1.personal.opinion.of.this.Deodorant', axis=1, inplace=True)
```

```
dados['q8.12'].fillna(dados['q8.12'].median(), inplace=True)
dados['q8.7'].fillna(dados['q8.7'].median(), inplace=True)
```

```
#Contabilizando dados faltantes
faltantes = dados.isnull().sum()
faltantes_porcentual = dados.isnull().sum() / len(dados['Product.ID']) * 100
print(faltantes_porcentual)
```

```
#Separando variaveis entre preditoras e target
y = dados['Instant.Liking']
x = dados.drop('Instant.Liking', axis=1)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

#separando os dados em folds:
stratified = StratifiedKFold(n_splits=5)

#Criando o modelo:
modelo = LogisticRegression(max_iter=1000)
resultado = cross_val_score(modelo,x,y,cv=stratified)

#Imprimindo a acurácia:
print(resultado.mean())
```
