```
import pandas as pd

dados = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Modulo-1/Regressao-Linear/datasets_14872_228180_Admission_Predict.csv')
dados.head()
```

```
dados.drop('Serial No.', axis=1, inplace=True)
```

```
#Separando variaveis predidoras e variavel target
x = dados.drop('Chance of Admit ', axis = 1)
y = dados['Chance of Admit ']
```

```
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
```

```
modelo = LinearRegression()
kfold = KFold(n_splits=5)
resultado = cross_val_score(modelo,x,y,cv = kfold) 
```

```
print(resultado.mean())
```
