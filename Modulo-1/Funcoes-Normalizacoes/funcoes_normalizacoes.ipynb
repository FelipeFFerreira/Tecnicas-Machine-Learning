```
from sklearn.preprocessing import MinMaxScaler
#MinMaxScaler
'''X_std = (X - Xmin) / (Xmax - Xmin)
x_scaled = x_std * (max - mim) + min'''
X =[[4, 1, 2, 2], [1, 3, 9, 3], [5, 7, 5, 1]]
normalizador = MinMaxScaler(feature_range = (0 , 1))
print(normalizador.fit_transform(X))
```

```
from sklearn.preprocessing import StandardScaler
#StandardScaler
'''z = (x - u) / s
Onde u é a media e "s" é o desvio padrao(standard deviation)'''
X =[[4, 1, 2, 2], [1, 3, 9, 3], [5, 7, 5, 1]]
normalizador = StandardScaler()
print(normalizador.fit_transform(X))
```

```
from sklearn.preprocessing import MaxAbsScaler
#MaxAbsScaler
'''
Normaliza os dados dividindo cada elemento pelo maior valor do conjunto
X'=X/M(onde M é o valor maximo)'''
X =[[4, 1, 2, 2], [1, 3, 9, 3], [5, 7, 5, 1]]
normalizador = MaxAbsScaler()
print(normalizador.fit_transform(X))
```

```
#Normalize
from sklearn.preprocessing import normalize
'''
Realiza a normalização de *cada linha da matriz* ( o calculo é feito linha por linha em vez de coluna por coluna) *axis = 0
normaliza por coluna. Possui 3 parametros possiveis, l1, l2 e max
L1 = ||x||1 = Soma(xi)
L2 = ||x||1 = Soma(raiz((xi²))
Max z = Max xi
'''
X =[[4, 1, 2, 2], [1, 3, 9, 3], [5, 7, 5, 1]]
normalizador = normalize(X, norm = 'max', axis=0)
print(normalizador)
```
