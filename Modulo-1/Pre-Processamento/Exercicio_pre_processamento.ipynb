Tramento com dados NaN. Iremos calcular a porcentagem em cada variavel

```
dados_NaN_percentual = (dados.isnull().sum() / len(dados['OSEBuildingID'])) * 100
print(dados_NaN_percentual)
```

**Trocando todas as Variavel 'ENERGYSTARScore' NaN pela sua mediana**

```
 dados['ENERGYSTARScore'] = dados['ENERGYSTARScore'].fillna(dados['ENERGYSTARScore'].median())
```
