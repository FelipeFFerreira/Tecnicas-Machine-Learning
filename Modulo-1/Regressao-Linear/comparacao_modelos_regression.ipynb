```
import pandas as pd
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', 21)
dados = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Modulo-1/Regressao-Linear/kc_house_data.csv')
dados.drop('id', axis = 1, inplace = True)
dados.drop('date', axis = 1, inplace = True)
dados.drop('zipcode', axis = 1, inplace = True)
dados.drop('lat', axis = 1, inplace = True)
dados.drop('long', axis = 1, inplace = True)

y = dados['price']
x = dados.drop(['price'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state=58)

dados.head()
```

**FUNCAO DE COMPARAÃ‡AO DE MODELOS REGRESSOES LINEARES**

```
def modelos_regression_linear(X_train, X_test, y_train, y_test):
  from sklearn.linear_model import LinearRegression
  from sklearn.linear_model import Ridge
  from sklearn.linear_model import Lasso
  from sklearn.linear_model import ElasticNet

  modelo = LinearRegression()
  modelo.fit(X_train, y_train)
  resultado = modelo.score(X_test, y_test)

  modeloRidge = Ridge()
  modeloRidge.fit(X_train, y_train)
  resultadoRidge = modeloRidge.score(X_test, y_test)

  modeloLasso = Lasso()
  modeloLasso.fit(X_train, y_train)
  resultadoLasso = modeloLasso.score(X_test, y_test)

  modeloElasticNet = ElasticNet()
  modeloElasticNet.fit(X_train, y_train)
  resultadoElasticNet = modeloElasticNet.score(X_test, y_test)
  print('\nModelo LinearRegression= ', resultado, ' Modelo Ridge= ',  resultadoRidge,  'Modelo Lasso= ', resultadoLasso,' Modelo Elastic Net= ', resultadoElasticNet)
  return [['Regressao linear', resultado ], ['Modelo Ridge', resultadoRidge], ['Modelo Lasso', resultadoLasso], ['Modelo Elastic Net', resultadoElasticNet]]
  
```

```
def melhor_modelo(resultados):
  i = 0
  maior = resultados[0]
  for x in resultados:
    if maior[1]  < x[1]:
      maior = x
  return maior

resultados = modelos_regression_linear(X_train, X_test, y_train, y_test)
best_model = melhor_modelo(resultados)
print('\n\nMelhor modelo encontrado: ', best_model[0], ' Com uma score de: ', best_model[1],'\n\n')
```
