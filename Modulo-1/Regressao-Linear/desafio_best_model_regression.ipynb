```
#Importando o modelo p/ dataframe pandas

import pandas as pd
from sklearn.model_selection import train_test_split

dados = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Modulo-1/Regressao-Linear/datasets_14872_228180_Admission_Predict.csv')
dados.head()
```

```
#Verificando se existi dados faltantes
dados_faltantes = dados.isnull().sum()
print(dados_faltantes)
```

```
#Excluindo Variavel desnecessaria
dados.drop('Serial No.', axis= 1, inplace=True)
```

```
#Definindo variaveis Preditoras e Target
x = dados.drop(['Chance of Admit '], axis=1)
y = dados['Chance of Admit ']
```

```
#Separando dados de treino e teste
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state=45)
```

```
#Aplicando modelos de Regressao Linear, Redge, Lasso e ElasticNet
#Comparando e validando best model with kfold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

def modelos_regression_linear(X_train, X_test, y_train, y_test):
  from sklearn.linear_model import LinearRegression
  from sklearn.linear_model import Ridge
  from sklearn.linear_model import Lasso
  from sklearn.linear_model import ElasticNet

  modelo = LinearRegression()
  modelo.fit(X_train, y_train)
  resultado = modelo.score(X_test, y_test)

  modeloRidge = Ridge()
  modeloRidge.fit(X_train, y_train)
  resultadoRidge = modeloRidge.score(X_test, y_test)

  modeloLasso = Lasso()
  modeloLasso.fit(X_train, y_train)
  resultadoLasso = modeloLasso.score(X_test, y_test)

  modeloElasticNet = ElasticNet()
  modeloElasticNet.fit(X_train, y_train)
  resultadoElasticNet = modeloElasticNet.score(X_test, y_test)
  print('\nModelo LinearRegression= ', resultado, ' Modelo Ridge= ',  resultadoRidge,  'Modelo Lasso= ', resultadoLasso,' Modelo Elastic Net= ', resultadoElasticNet)
  
  kfold = KFold(n_splits=10)
  resultado_linear_kfold = cross_val_score(modelo,x,y,cv = kfold)
  resultado_ridge_kfold = cross_val_score(modeloRidge,x,y,cv = kfold)
  resultado_lasso_kfold = cross_val_score(modeloLasso,x,y,cv = kfold)
  resultado_elastic_kfold = cross_val_score(modeloElasticNet,x,y,cv = kfold)
  return [['Regressao linear', resultado_linear_kfold.mean() ], ['Modelo Ridge', resultado_ridge_kfold.mean()], ['Modelo Lasso', resultado_lasso_kfold.mean()], ['Modelo Elastic Net', resultado_elastic_kfold.mean()]]
  
```

```
#Determina qual melhor modelo linear que representa os dados
def melhor_modelo(resultados):
  i = 0
  maior = resultados[0]
  for x in resultados:
    if maior[1]  < x[1]:
      maior = x
  return maior

resultados = modelos_regression_linear(X_train, X_test, y_train, y_test)
best_model = melhor_modelo(resultados)
print('\n\nMelhor modelo encontrado: ', best_model[0], ' Com uma score de: ', best_model[1],'\n\n')
```
