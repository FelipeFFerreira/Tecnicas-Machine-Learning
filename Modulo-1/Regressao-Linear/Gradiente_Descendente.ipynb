```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

def derivadam(x,y,m,b):
    derivada_m = 0
    for _x, _y in zip(x, y):
        derivada_m += 2 * ((m * _x) + b - _y) * _x
    return derivada_m

def derivadab(x,y,m,b):
    derivada_b = 0
    for _x, _y in zip(x, y):
        derivada_b += 2 * ((m * _x) + b - _y)
    return derivada_b
 
def regression_linear(m,x,b):
    result = []
    for p in x:
        result.append(m*p + b)
    return result

def principal():
    x, y = make_regression(n_samples=200, n_features=1, noise=30)
    modelo = LinearRegression()
    modelo.fit(x,y)
    m = 7
    b = 1
    L = 0.0001 #Taxa de aprendizagem
    for i in range(1000):
        m -= derivadam(x,y,m,b)*L
        b -= derivadab(x,y,m,b)*L
    #y_pred = regression_linear(m,x,b)
    xreg = np.arange(-3, 3, 1)
    plt.scatter(x, y)
    plt.plot(xreg, m*xreg + b, color = 'red')
    #plt.plot(xreg, modelo.coef_*xreg + modelo.intercept_, color = 'yellow')
    plt.show()
    print("modelo b=",modelo.intercept_)
    print("calc b=", b)
    print("modelo m=", modelo.coef_)
    print("modelo m=", m)
principal()
```
